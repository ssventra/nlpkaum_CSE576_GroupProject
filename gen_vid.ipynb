```python
items = {}
present_class = None
# previous_class = None
for idx, node in enumerate(x[1]['nodes']):
    if 'ROO_' in node['prefab_name']:
        present_class = node['class_name']
        items[present_class] = [idx,0]
    else:
        if present_class is None:
            continue
        if items[present_class] is not None:
            items[present_class][1] = idx
```
